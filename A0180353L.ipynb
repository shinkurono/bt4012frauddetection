# Oversampling (SMOTE)

SMOTE was chosen for oversampling, and reduces the problem of overfitting posed by the original unbalanced dataset

```python
#pip install --U imblearn
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.metrics import roc_auc_score
```

```python
sm = SMOTE(random_state=47)
X_train_os, y_train_os = sm.fit_sample(X_train, y_train)
```

```python
X_train_os.shape
```

Here we reshape the row into individual 20 x 20 arrays

```python
X_train = np.asarray(X_train_os).reshape(91804,20,20,1)
y_train = np.asarray(y_train_os)
```

```python
X_test.shape
```

```python
X_test = np.asarray(X_test).reshape(21642,20,20,1)
y_test = np.asarray(y_test)
```

```python
pd.Series(y_train).value_counts().sort_index()
```

After which, we normalise the data by dividing it by 255, which is the max in a RGB spectrum of light

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

# CNN

Model is built here along with a defined AUC scoring function as there is no predefined AUC scoring funciton.

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from tensorflow.keras.optimizers import Adam, SGD, RMSprop, Adamax

def auroc(y_true, y_pred):
    return tf.py_function(roc_auc_score, (y_true, y_pred), tf.double)

stride_size = (1, 1)
num_filters = 4

cnn_model = Sequential()
# feature extraction layer 0: convolution
for kernel_size in [(2,2),(3,3),(4,4), (5,5)]:
    cnn_model.add(Conv2D(num_filters*4, kernel_size=kernel_size,
                 activation='relu',
                 input_shape=(20, 20, 1),
                 padding = 'same'))
    # feature extraction layer 1: max pooling

cnn_model.add(MaxPooling2D(pool_size=(2, 2)))
cnn_model.add(Dropout(0.5))
# classification layer 2: flattening
cnn_model.add(Flatten())
# classification layer 3: dense non-linear transformation
cnn_model.add(Dense(100, activation='relu'))
# classification layer 4: output label probability
cnn_model.add(Dense(1, activation='sigmoid'))

cnn_model.compile(loss='binary_crossentropy',
                  optimizer=Adam(lr=0.0001),
                  metrics=[auroc])

```

```python
import tensorflow as tf
```

```python
cnn_model.fit(x=X_train,y=y_train,epochs=50)
```

```python
print(roc_auc_score(y_train, cnn_model.predict(X_train)))
print(roc_auc_score(y_test, cnn_model.predict(X_test)))
```
